```python
dataset.shape[0]
```

```python
# percentage null vale
# I use percentage formula
(dataset.isnull().sum()/dataset.shape[0])*100
```

```python
# check not null valuer
dataset.notnull().sum()
```

```python
# Total not null value
dataset.notnull().sum().sum()
```
